# Exercise 2: Decision-Making Under Certainty

```python
import numpy as np
```

```python
def simulate_decision(num_simulations, p_success, reward_success, reward_failure):
    outcomes = []
    for _ in range(num_simulations):
      if np.random.rand() < p_success:
          outcomes.append(reward_success)
      else:
          outcomes.append(reward_failure)
    return np.mean(outcomes)


p_success = 0.7
reward_success = 1000
reward_failure = -500

average_outcome = simulate_decision(1000, p_success, reward_success, reward_failure)
print(f"Expected value of decision: {average_outcome}")
```

**Assessment Task 2: Bayesian Inference**

```python
def bayesian_inference(prior_disease, sensitivity, false_positive_rate):
    # Likelihood of a positive test result given disease
    p_e_given_h = sensitivity

    # Probability of not having the disease
    p_not_h = 1 - prior_disease

    # Likelihood of a positive test result given no disease
    p_e_given_not_h = false_positive_rate

    # Total probability of a positive test result
    p_e = (p_e_given_h * prior_disease) + (p_e_given_not_h * p_not_h)

    # Posterior probability using Bayes' theorem
    posterior = (p_e_given_h * prior_disease) / p_e

    return posterior

# Given values
prior_disease = 0.01  # 1% prevalence
sensitivity = 0.90    # 90% true positive rate
false_positive_rate = 0.05  # 5% false positive rate

# Calculate posterior probability
posterior_probability = bayesian_inference(prior_disease, sensitivity, false_positive_rate)
print(f"The probability that someone who tested positive actually has the disease is: {posterior_probability:.4f}")
```

**Assessment Task 3: Decision-Making Under Uncertainty**

```python
import random

def investment_simulation(iterations, probability_success, reward, penalty):

    total_return = 0

    for _ in range(iterations):

        random_number = random.random()


        if random_number < probability_success:
            total_return += reward
        else:
            total_return += penalty


    average_return = total_return / iterations

    return average_return

# Given parameters
iterations = 1000
probability_success = 0.7
reward = 1000
penalty = -500

# Run the investment simulation
average_return = investment_simulation(iterations, probability_success, reward, penalty)
print(f"The average return after {iterations} iterations is: ${average_return:.2f}")
```

**Assessment Task 4: Real-World Scenario**


```python
import numpy as np
import matplotlib.pyplot as plt


P_rain = 0.3
P_no_rain = 1 - P_rain


P_humidity_given_rain = 0.9
P_humidity_given_no_rain = 0.6


P_cloud_cover_given_rain = 0.8
P_cloud_cover_given_no_rain = 0.5

def calculate_probability_of_rain(humidity, cloud_cover):

    # Calculate likelihoods
    P_humidity = (humidity * P_humidity_given_rain +
                   (1 - humidity) * P_humidity_given_no_rain)

    P_cloud_cover = (cloud_cover * P_cloud_cover_given_rain +
                     (1 - cloud_cover) * P_cloud_cover_given_no_rain)

    # Calculate joint probabilities
    P_humidity_and_cloud_cover_given_rain = (
        P_humidity_given_rain * P_cloud_cover_given_rain
    )

    P_humidity_and_cloud_cover_given_no_rain = (
        P_humidity_given_no_rain * P_cloud_cover_given_no_rain
    )

    # Apply Bayes' theorem
    numerator = (P_humidity_and_cloud_cover_given_rain * P_rain)
    denominator = (
        numerator +
        (P_humidity_and_cloud_cover_given_no_rain * P_no_rain)
    )

    return numerator / denominator

# Interactive simulation
def interactive_simulation():
    humidity_values = np.linspace(0, 1, 100)
    cloud_cover_values = np.linspace(0, 1, 100)

    # Create a meshgrid for plotting
    H, C = np.meshgrid(humidity_values, cloud_cover_values)

    # Calculate probabilities for each combination of humidity and cloud cover
    probabilities_of_rain = np.array([
        [calculate_probability_of_rain(h, c) for h in humidity_values]
        for c in cloud_cover_values
    ])

    # Plotting the results
    plt.figure(figsize=(10, 6))
    plt.contourf(H, C, probabilities_of_rain, levels=20, cmap='viridis')
    plt.colorbar(label='Probability of Rain')

    plt.title('Probability of Rain based on Humidity and Cloud Cover')
    plt.xlabel('Humidity Level')
    plt.ylabel('Cloud Cover Level')

    plt.xlim(0, 1)
    plt.ylim(0, 1)

    plt.show()

# Run the interactive simulation
interactive_simulation()
```

```python
report = """
Effect of Humidity: Higher humidity generally increases the likelihood of rain since moist air is more conducive to precipitation.
Effect of Cloud Cover: Increased cloud cover also raises the probability of rain because clouds are necessary for precipitation to occur.
"""
print(report)
```

**Assessment Task 5: Probability Distributions Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm
```

```python
def visualize_binomial_distribution(n, p):

    x = np.arange(n + 1)
    pmf = binom.pmf(x, n, p)

    plt.figure(figsize=(8, 6))
    plt.bar(x, pmf)
    plt.title(f"Binomial Distribution (n={n}, p={p:.2f})")
    plt.xlabel("Number of Successes")
    plt.ylabel("Probability")
    plt.show()

def visualize_normal_distribution(mean, std_dev, lower, upper):

    x = np.linspace(lower, upper, 100)
    pdf = norm.pdf(x, mean, std_dev)

    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf)
    plt.title(f"Normal Distribution (μ={mean:.2f}, σ={std_dev:.2f})")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.show()

```

```python
def simulate_coin_flips(n, p):

    return binom.rvs(n, p)

# Simulate 1000 coin flips with a 0.5 probability of heads
n = 1000
p = 0.5
num_heads = simulate_coin_flips(n, p)

print(f"Number of heads in {n} coin flips: {num_heads}")

visualize_binomial_distribution(n=10, p=0.5)
visualize_normal_distribution(mean=0, std_dev=1, lower=-3, upper=3)
```
